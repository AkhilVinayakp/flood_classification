```python
import numpy as np
import scipy.fftpack as fftpack
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('../data/filt_data.csv')
```

```python
data
```

```python
gx = np.array(data[0:50]['gx'])
```

```python
gx
```

```python
gxfft = fftpack.fft(gx)
```

```python
gxfft
```

```python
amplitudes = np.abs(gxfft)
```

```python
power = amplitudes**2
```

```python
gx.size
```

```python
freq = fftpack.fftfreq(gx.size,d = .1)
```

```python
amplitudes
```

```python
freq
```

```python
plt.figure(figsize=(20,20))
plt.stem(freq,amplitudes,use_line_collection=True)
```

```python
plt.figure(figsize=(20,20))
plt.stem(amplitudes,use_line_collection=True)
```
